Define a class Car with attributes and a method to display details.

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display(self):
        print("Brand:", self.brand)
        print("Model:", self.model)

car = Car("Toyota", "Corolla")
car.display()
```

Explain the purpose of the __init__ method.

The init() method is called automatically when an object is created. It is known as the constructor. It is used to initialize the object's attributes.

```python
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Rahul")
print(s.name)
```

Create a subclass ElectricCar that inherits from Car.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, brand, battery):
        super().__init__(brand)
        self.battery = battery

e = ElectricCar("Tesla", "75 kWh")
print(e.brand)
print(e.battery)
```

Demonstrate "Method Overriding" in inheritance.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

Explain "Encapsulation" by creating private attributes using __.

Encapsulation means hiding data from direct access using private attributes.

```python
class Student:
    def __init__(self):
        self.__marks = 90

    def show_marks(self):
        print(self.__marks)

s = Student()
s.show_marks()
```

Use @property to create getter and setter methods.

```python
class Student:
    def __init__(self):
        self.__marks = 0

    @property
    def marks(self):
        return self.__marks

    @marks.setter
    def marks(self, value):
        self.__marks = value

s = Student()
s.marks = 95
print(s.marks)
```

What is "Multiple Inheritance"? Provide a small code structure.

Multiple Inheritance means a class inherits from more than one parent class.

```python
class Father:
    def skill1(self):
        print("Driving")

class Mother:
    def skill2(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skill1()
c.skill2()
```

Create a class method using the @classmethod decorator.

```python
class Student:
    school = "ABC School"

    @classmethod
    def show_school(cls):
        print(cls.school)

Student.show_school()
```

Explain the difference between a class method and a static method.

Class Method:
- Uses @classmethod
- Takes cls as the first parameter.
- Can access and modify class variables.

Static Method:
- Uses @staticmethod
- Takes no special parameter (self or cls).
- Cannot directly access class or instance variables unless passed explicitly.

```python
class Demo:
    @classmethod
    def class_method(cls):
        print("This is a class method")

    @staticmethod
    def static_method():
        print("This is a static method")

Demo.class_method()
Demo.static_method()
```

Write a try-except-finally block to handle a ZeroDivisionError.

```python
try:
    a = 10
    b = 0
    c = a / b
    print(c)

except ZeroDivisionError:
    print("Cannot divide by zero.")

finally:
    print("Program execution completed.")
```
